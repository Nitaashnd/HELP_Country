## **Penjelasan kolom fitur:**


**Negara** : Nama negara

**Kematian_anak**: Kematian anak di bawah usia 5 tahun per 1000 kelahiran

**Ekspor** : Ekspor barang dan jasa perkapita

**Kesehatan**: Total pengeluaran kesehatan perkapita

**Impor**: Impor barang dan jasa perkapita

**Pendapatan**: Penghasilan bersih perorang

**Inflasi**: Pengukuran tingkat pertumbuhan tahunan dari Total GDP 

**Harapan_hidup**: Jumlah tahun rata-rata seorang anak yang baru lahir akan hidup jika pola kematian saat ini tetap sama

**Jumlah_fertiliti**: Jumlah anak yang akan lahir dari setiap wanita jika tingkat kesuburan usia saat ini tetap sama

**GDPperkapita**: GDP per kapita. Dihitung sebagai Total GDP dibagi dengan total populasi.

**Handle outlier**

```python
# boxplot sebelum dibuang outliernya
plt.figure(figsize=(10,8))

for i in enumerate(df.describe().columns):
  plt.subplot(3, 3, i[0]+1)
  sns.boxplot(df[i[1]])
  plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

dari boxplot diatas, dapat diketahui bahwa setiap kolom mempunyai outlier, namun outlier tidak akan dihilangkan pada data ini karena akan berdampak pada peringkat negara yang membutuhkan bantuan keuangan dari HELP International. Oleh karena itu, digunakan pendekatan lain dengan membatasi outlier karena goal dari analisis ini adalah untuk mencari daftar negara yang membutuhkan bantuan keuangan. Maka, digunakan pembatasan sebagian kecil outlier. untuk meminimalkan bias, pembatasan akan didasarkan pada persentil ke-99.

```python
# Capture outliers
cap_features = ['Ekspor', 'Impor', 'Kesehatan' , 'Jumlah_fertiliti', 
                'Harapan_hidup', 'Pendapatan' ,'Inflasi' , 'GDPperkapita']
dfnew = df.copy()
for col in cap_features:
    q4 = dfnew[col].quantile(0.99)
    dfnew.loc[dfnew[col] >= q4, col] = q4
```

```python
# boxplot sesudah di handle outliernya
plt.figure(figsize=(10,8))

for i in enumerate(dfnew.describe().columns):
  plt.subplot(3, 3, i[0]+1)
  sns.boxplot(dfnew[i[1]])
  plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

```python
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
```

```python
#drop data non numerik agar dapat rescale data
dfnew2 = dfnew.drop(columns='Negara') #pakai yang di handle outliernya

display(dfnew2)
```

```python
feature = dfnew2
```

```python
scaler = StandardScaler()
ft_standardize = scaler.fit_transform(feature)
#agar tidak ada rentang yang terlalu jauh atributnya
```

```python
#Elbow method untuk melihat jumlah n-cluster yang akurat
def elbowMethod(data, k_min=2, k_max= 10):
    wcss = [] 
    k_range = range(k_min, k_max + 1)

    for i in k_range:
      kmeans_test = KMeans(n_clusters = i, random_state = 42, init = 'k-means++')
      kmeans_test.fit(data)
      wcss.append(kmeans_test.inertia_)

    fig, ax = plt.subplots(figsize=(15,8))
    ax.plot(k_range, wcss, marker='o')

    for i, value in enumerate(wcss):
        ax.text(i+2.15, value-0.005, round(value,3))

    plt.axvline(x = 4, color = 'r')
    plt.title('The Elbow Method')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS')
    plt.show()
```

```python
print('Elbow Method')
print()
elbowMethod(ft_standardize)
```

n = 4

```python
# Clustering with n_cluster 4
kmeans = KMeans(n_clusters = 4, random_state = 42).fit(ft_standardize)
labels = kmeans.labels_
```

```python
# banyak negara dalam cluster
print('klasifikasi berjumlah 4 macam')
print()
dfnew['Cluster'] = labels
print('cluster dan jumlah negara di dalamnya :')
display(dfnew.Cluster.value_counts(ascending=True))
print()
display(dfnew.head())
```

```python
# Display mean statistic of each columns after 4-clustering to represents centroids
analysis_res = dfnew.groupby('Cluster').agg({'mean'})
analysis_res['Jumlah_negara'] = dfnew.groupby('Cluster')['Negara'].count()
display(analysis_res)
```

```python
# Visualize mean statistic of each columns after 4-clustering
plt.figure(figsize=(10, 8))
for i, var in enumerate(dfnew2.columns):
    plt.subplot(3,3,i+1)
    sns.barplot(x = analysis_res.reset_index().Cluster, y = analysis_res[var]['mean'])
    plt.ylabel(var, fontsize=15)
    plt.xlabel('Cluster', fontsize=11)
plt.tight_layout()    
plt.show() 
```
